```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('ANZ synthesised transaction dataset.xlsx',encoding='sys.getfilesystemencoding()')
df.head()
```

```python
print(df.columns)
print(df.shape)
```

```python
# Confirmation for 100 unique accounts
df["account"].nunique()
```

```python
# Removing useless columns
df_clean = df[["status","card_present_flag","balance","date",
                   "gender","age","merchant_suburb","merchant_state",
                   "amount","customer_id","movement"]]
df_clean["date"] = pd.to_datetime(df_clean["date"])
df_clean.head()
```

```python
# Transactions made each day
df_clean["date"].value_counts()
```

```python
# Transactions made by each customer
df_clean["customer_id"].value_counts()
```

```python
# Transaction volume each day
df_date_count=df_clean.groupby("date").count()
trans_vol = df_date_count["customer_id"].mean()
n_points = len(df_date_count.index)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_date_count.index, df_date_count["customer_id"], c="black", label = "Customer ID")
plt.plot(df_date_count.index, np.linspace(trans_vol,trans_vol,n_points), c="red", label = "Mean transaction volume")
plt.title("Transactions Volume vs. Date")
plt.xlabel("Date")
plt.ylabel("Number of customers")
plt.legend()
plt.tight_layout()
```

```python
# Mean transaction amount each day
df_date_mean=df_clean.groupby("date").mean()
trans_amt = df_date_mean["amount"].mean()
n_points = len(df_date_count.index)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_date_count.index, df_date_mean["amount"], c="black", label = "Amount")
plt.plot(df_date_count.index, np.linspace(trans_amt,trans_amt,n_points), c="red", label = "Overall mean transaction amount")
plt.title("Mean Transaction Amount vs. Date")
plt.xlabel("Date")
plt.ylabel("Amount($)")
plt.legend()
plt.tight_layout()
```

```python
months = []
for date in df_clean["date"]:
    if date.month == 8:
        months.append("August")
    elif date.month == 9:
        months.append("September")
    elif date.month == 10:
        months.append("October")

df_clean["Months"] = months
df_clean["Months"].head()
```

```python
# mean customer balance and payment by age
df_cus_aug = df_clean[df_clean["Months"] == "August"].groupby("customer_id").mean()
df_gen_aug = df_clean[df_clean["Months"] == "August"].groupby("gender").mean()
```

```python
mean_f_bal_aug = df_gen_aug["balance"].iloc[0]
mean_m_bal_aug = df_gen_aug["balance"].iloc[1]
n_points = len(df_cus_aug["age"])
```

```python
plt.scatter(df_cus_aug["age"], df_cus_aug["balance"], c="black", label="Balance")
plt.plot(df_cus_aug["age"], np.linspace(mean_f_bal_aug,mean_f_bal_aug,n_points), c="red", label = "Mean Female balance")
plt.plot(df_cus_aug["age"], np.linspace(mean_m_bal_aug,mean_m_bal_aug,n_points), c="blue", label = "Mean Male balance")
plt.title("Customer Balance vs. Age for August")
plt.xlabel("Age(years)")
plt.ylabel("Balance($)")
plt.legend()
plt.tight_layout()
```

```python
mean_f_amt_aug = df_gen_aug["amount"].iloc[0]
mean_m_amt_aug = df_gen_aug["amount"].iloc[1]
```

```python
plt.scatter(df_cus_aug["age"], df_cus_aug["amount"], c="black", label="Amount")
plt.plot(df_cus_aug["age"], np.linspace(mean_f_amt_aug,mean_f_amt_aug,n_points), c="red", label = "Mean female amount")
plt.plot(df_cus_aug["age"], np.linspace(mean_m_amt_aug,mean_m_amt_aug,n_points), c="blue", label = "Mean male amount")
plt.title("Customer Mean Payment Amount vs. Age for August")
plt.xlabel("Age(years)")
plt.ylabel("Amount($)")
plt.legend()
plt.tight_layout()
```
